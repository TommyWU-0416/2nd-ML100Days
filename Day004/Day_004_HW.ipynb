### Series欄位
- 透過Series會將資料型別都轉成相同的
- 指定index(若未指定，將會是0, 1, 2, 3...

```python
s1 = pd.Series([1, 3, 6, np.nan, 44, 1])
print(type(s1))
print(s1)
```

```python
s2 = pd.Series([1, 4, 6], index=['a', 'b', 'c'])
print(type(s2))
s2
```

### DataFrame 表格
- random.randn產生隨機數字
- info() 資料表大小、欄位數量、資料型態

```python
df = pd.DataFrame(np.random.randn(6, 4), columns=['a', 'b', 'c', 'd'])
print(type(df))
df
```
